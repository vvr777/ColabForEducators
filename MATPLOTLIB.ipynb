### <font color='steelblue'>Crear y visualizar un gráfico

Hay dos funciones básicas para la representación de gráficos: `plot()` y `show()`. Para utilizarlas, tendremos que utilizar el prefijo `plt.`.

La primera, `plt.plot()`, recibe en un gráfico bidimensional, tres argumentos: los valores del eje $X$, los valores del eje $Y$, cada uno de ellos dispuestos en listas, y en el tercer argumento las especificaciones sobre cómo queremos visualizar las líneas y los puntos (colores, tipo de líneas, tipos de puntos, ...).

La función `plt.show()` se utiliza  al final, después de definir el gráfico, para visualizarlo.

En el ejemplo a continuación representamos una parábola. Para ello introducimos una secuencia de valores $x$, y para cada uno de ellos calculamos su cuadrado, $y=x^2$. Guardamos ambas secuencias en sendas listas.

````python
# valores de "x"
x = list(range(6))

# creamos "y" vacía y la rellenamos con los cuadrados de x
y = []
for i in x:
  # vamos añadiendo cada valor de "x" al cuadrado
  y.append(i**2)

# visualizamos ambas listas
print(x)
print(y)
````

Sin preocuparnos sobre especificaciones de estilo, creamos el gráfico solo con las listas 'x' e 'y'. Aparecen las especificaciones básicas: una línea continua azul que une los puntos introducidos $(x,y)$, sin marcadores para los puntos.

````python
plt.plot(x, y)
plt.show()
````

Fíjate, que si no utilizas `plt.show()` aparece un mensaje, por lo que es recomendable escribirlo para dejar todo más ordenado. Pruébalo a continuación.

````python
plt.plot(x, y)
````

## <font color='steelblue'> 2. Especificaciones básicas


Cuando queremos configurar los estilos de un gráfico, podemos hacerlo de dos modos:

1. utilizar el tercer argumento en la función `plot()` para introducir, en *formato corto* las especificaciones pertinentes entrecomilladas y en un orden concreto;
1. utilizar los argumentos correspondientes de la función `plot()` en *formato largo*,  y en el orden que se desee.

### <font color='steelblue'>Formato corto para líneas y marcadores

Cuando utilizamos el formato corto, incluimos, básicamente, un carácter para reconocer cada una de las siguientes características, y siempre en el mismo orden: *color, tipo de marcador y tipo de línea*.


Así, si especificamos en formato corto como tercer argumento `'g^--'`, estaremos pidiendo un gráfico de color verde ('g' de *green*), con triángulos como símbolos para los puntos ('^') y con líneas discontinuas (`'--'`). Veamos cómo queda el gráfico anterior con estas especificaciones.

````python
plt.plot(x, y, 'g^--')
plt.show()
````

Las posibilidades para las características básicas son:
- COLOR. Se suele utilizar la inicial del color en inglés. Por ejemplo: 'r' hace referencia a rojo (*red*), 'b' a azul (*blue*), 'y' a amarillo (*yellow*). El color negro se identifica por 'k'. Si no se especifica nada, por defecto los puntos y líneas se dibujan en azul.

- PUNTOS o marcadores. Un punto redondo se representa con 'o', un triángulo con '^', 's' si queremos un cuadrado (*square*), '*' si queremos un asterisco, 'd' para un diamante y 'D' para rombos.  Si no especificamos nada, por defecto no se pintan los puntos.

- TIPOS DE LÍNEA. Tenemos la opción '-' para líneas continuas, '--' para discontinuas, ':' para líneas punteadas, y '-.' para líneas con líneas y puntos.

Si no se especifica nada para las líneas, pero sí para los marcadores, aparecen solo los marcadores, sin líneas, salvo cuando el color es negro ('k'). Por contra, si solo definimos el color, aparecerá una línea continua sin marcadores.

Veamos a continuación algunos ejemplos más.

````python
# gráfico en negro, marcadores con círculos y línea punteada uniendo puntos
plt.plot(x, y, 'ko:')
plt.show()
````

````python
# gráfico en rojo, marcadores con diamantes y línea con puntos y rayas
plt.plot(x, y, 'rd-.')
plt.show()
````

````python
# solo marcadores (puntos) en amarillo y con forma de cuadrado
plt.plot(x, y, 'ys')
plt.show()
````

### <font color='steelblue'>Formato largo para líneas y marcadores

La función `plt.plot()` dispone de argumentos específicos para controlar las tres características configurables con el formato corto, así como algunas otras. En concreto, el color, el tipo de marcador para los puntos y el tipo de línea lo especificamos, en formato largo, con los argumentos `color`, `marker` y `linestyle`.

Además, algunos de los argumentos de `plot()` tienen un nombre largo -como `color`-, y una abreviatura -en este caso `c`-, que indicamos a continuación al describirlos:

* `color/c`, donde especificamos, entrecomillado, el nombre del color en inglés (`'green'`, `'red'`, `'black'`, `'yellow'`, `'purple'`, `'grey'`, etc.), o bien el *código hexadecimal* inicializado por una almohadilla (por ejemplo, `color='#bc2a30'`); en [htmlcolorcodes](https://htmlcolorcodes.com/es/) tienes muchas posibilidades para elegir.
* `marker`. Los tipos de marcador se especifican igual que en el formato corto: `'s'`, `'o'`, `'^'`, `'*'`, u otros que puedes encontrar en la [web de matplotlib](https://matplotlib.org/stable/api/markers_api.html). Este argumento no tiene abreviatura.
* `linestyle/ls`. Para el tipo de línea también se usan los mismos caracteres que en formato corto: `'-'`, `'--'`, `':'` y `'-.'`. Si no queremos que haya líneas, el valor de este argumento estará vacío (`linestyle=''`).


Veamos algunos ejemplos con los argumentos en formato largo.

````python
plt.plot(x, y, color='black', marker='o', linestyle=':')
# equivalente a plt.plot(x, y, c='black', marker='o', ls=':')
plt.show()
````

````python
plt.plot(x, y, color='red', marker='d', linestyle='-.')
# equivalente a plt.plot(x, y, c='red', marker='d', ls='-.')
plt.show()
````

### <font color='steelblue'>Otras especificaciones

Además de las características básicas como el color, el tipo de marcador y de línea, existen otras características que podemos configurar a través de argumentos de la función `plot()`. La mayoría de estos argumentos tiene también una abreviatura que podemos utilizar indistintamente, y que indicamos a continuación:

* `alpha` para controlar la opacidad (transparencia) del gráfico, con un valor entre 0 (transparencia total y no se ve nada) y 1 (opacidad total). Este argumento no tiene abreviatura.
* `linewidth/lw` para el grosor de la línea, indicado por un número mayor que cero (el cero no pinta línea).
* `markersize/ms` controla el tamaño de los marcadores, siendo 6 el tamaño por defecto. Se especifica con números positivos.
* `markerfacecolor/mfc` permite especificar el color de relleno de los marcadores, utilizando, igual que en `color`, el nombre del color en inglés o el código hexadecimal con almohadilla.
* `markeredgewidth/mew` controla el ancho del borde del marcador con un número. Si no se indica nada, su tamaño es 1. Si iguala o supera el tamaño del marcador, lo tapa.
* `markeredgecolor/mec` indica el color del borde del marcador,con las mismas opciones que `color`.


A continuación vamos añadiendo argumentos al ejemplo que venimos trabajando.

````python
# color, marcador, ancho de línea y opacidad
plt.plot(x, y, color='black', marker='s', linewidth=2,alpha=0.5)
# equivalente a plt.plot(x, y, c='black', marker='s', lw=2,alpha=0.5)
plt.show()
````

````python
# tamaño del marcador
plt.plot(x, y, color='black', marker='s', linewidth=2, markersize=9)
# equivalente a plt.plot(x, y, c='black', marker='s', lw=2, ms=9)
plt.show()
````

````python
# Color de relleno del marcador
plt.plot(x, y, color='black', marker='s', linewidth=2, markersize=9,
         markerfacecolor='yellow')
# equivalente a plt.plot(x, y, c='black', marker='s', lw=2, ms=9,mfc='yellow')
plt.show()
````

````python
# Ancho del borde del marcador
plt.plot(x, y, color='black', marker='s', linewidth=2, markersize=9,
         markerfacecolor='yellow', markeredgewidth=3)
# equivalente a plt.plot(x, y, c='black', marker='s', lw=2, ms=9,mfc='yellow', mew=3)
plt.show()
````

````python
# Color del borde del marcador
plt.plot(x, y, color='black', marker='s', linewidth=2, markersize=9,
         markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='blue')
# equivalente a plt.plot(x, y, c='black', marker='s', lw=2, ms=9,mfc='yellow', mew=3, mec='blue')
plt.show()
````

## <font color='steelblue'> 3. Varias representaciones en un gráfico

Hay veces que queremos dibujar dos funciones diferentes en un mismo gráfico, generalmente para compararlas. Para ello basta con escribir dos veces la función `plot`, una con cada una de las funciones a representar, y solicitar después la visualización con la función `show`.

En nuestro ejemplo de representación de parábolas, vamos a representar el rango que va entre -5 y 5 para la variable `X`. Definimos pues `x` como una secuencia en este rango. Creamos a continuación las variables `y1` e `y2`, que recogen respectivamente los valores de la función $y=x^2 - 2$ y de $y=-x^2 + 2$ para los `x` dados.

````python
# valores de "x"
x = list(range(-5,6))

# creamos las "y"s vacías
y1 = []
y2 = []
for i in x:
  # añadimos los valores
  y1.append(i**2 - 2)
  y2.append(-i**2 + 2)

# vemos los valores
print(x)
print(y1)
print(y2)
````

Una vez hemos creado las dos funciones ya podemos representarlas en el mismo gráfico:

````python
import matplotlib.pyplot as plt
%matplotlib inline

# Función 1
plt.plot(x, y1)
# Función 2
plt.plot(x, y2)

# Mostramos el gráfico
plt.show()
````

Si no indicamos nada en los argumentos, la primera función se representa en color azul y la segunda en color naranja.

Vamos a personalizar ambas funciones, para diferenciarlas como deseemos.

````python
# Función 1 en rojo, marcadores rojos y línea continua
plt.plot(x, y1, 'ro-')
# Función 2 en negro, marcadores cuadrados y línea punteada
plt.plot(x, y2, 'ks:')

# Mostramos el gráfico
plt.show()
````

Por supuesto, podemos combinar el formato corto o largo en la especificación de los argumentos de `plot()`.

````python
# Función 1
plt.plot(x, y1, 'ro-')
# Función 2
plt.plot(x, y2, color='black', marker='s', linestyle=':', linewidth=4)

# Mostramos el gráfico
plt.show()
````

### <font color='steelblue'>Leyendas

Cuando representamos datos o funciones, es importante identificar qué representa cada color, forma o tamaño, y para ello utilizamos las leyendas. Las leyendas se generan con la función `legend()` del módulo `matplotlib.pyplot` (`plt`), antes de mostrar la figura (con `show()`) y después de dibujarlas (`plot()`).

Si utilizamos simplemente la función `legend()`sin especificar qué etiquetas queremos mostrar en la leyenda, obtendremos un *warning* avisándonos de ello. Veámoslo con el ejemplo anterior.

````python
# Función 1
plt.plot(x, y1, 'ro-')
# Función 2
plt.plot(x, y2, color='black', marker='s', linestyle=':', linewidth=4)
# Indicamos la leyenda
plt.legend()

# Mostramos el gráfico
plt.show()
````

Hemos pues de especificar las etiquetas que deseamos que aparezcan en la leyenda a través del argumento `label` en la función `plot()`. Cada función dibujada con `plot` mostrará en la leyenda la cadena de caracteres que hayamos definido en `label`.  

Con el ejemplo anterior, vamos a identificar las dos funciones representadas con las etiquetas 'y1' e 'y2'. Veámoslo.

````python
# Función 1
plt.plot(x, y1, 'ro-', label='y1')
# Función 2
plt.plot(x, y2, color='black', marker='s', linestyle=':', linewidth=4, label='y2')
# Indicamos la leyenda
plt.legend()

# Mostramos el gráfico
plt.show()
````

Cuando queremos que las etiquetas contengan símbolos matemáticos (por ejemplo, la definición de las funciones que representan), hemos de utilizar notación $\LaTeX$, escribiendo entre comillas la fórmula matemática encajada entre dos símbolos de dólar, `$...$`. Lo hacemos a continuación.

````python
# Función 1
plt.plot(x, y1, 'ro-', label='$x^2 - 2$')
# Función 2
plt.plot(x, y2, color='black', marker='s', linestyle=':', linewidth=4, label='$-x^2 + 2$')
# Indicamos la leyenda
plt.legend()

# Mostramos el gráfico
plt.show()
````

### <font color='steelblue'>Posición de la leyenda

Por defecto, la leyenda se suele ubicar, automáticamente, en algún lugar del gráfico en el que no interfiere con las líneas y puntos representados. Sin embargo, podemos indicar dónde queremos que aparezca. Para ello utilizaremos el argumento `loc=` dentro de la función `legend`. Los valores que acepta este argumento son: `'upper right'`, `'lower right'`, `'upper left'`, `'lower left'`, `'upper center'`, `'lower center'`, `'center left'`, `'center right'`, `'center'` y `'best'`. La opción `'best'` corresponde con la ubicación por defecto de `matplotlib`.


Para saber más sobre el posicionamiento de leyendas, se recomienda consultar la [ayuda de matplotlib](https://matplotlib.org/2.0.2/users/legend_guide.html).

Probemos algunas opciones de posicionamiento.

````python
# Función 1
plt.plot(x, y1, 'ro-', label='$x^2 - 2$')
# Función 2
plt.plot(x, y2, color='black', marker='s', linestyle=':', linewidth=4, label='$-x^2 + 2$')
# Colocamos la leyenda arriba a la derecha
plt.legend(loc='upper right')

# Mostramos el gráfico
plt.show()
````

````python
# Función 1
plt.plot(x, y1, 'ro-', label='$x^2 - 2$')
# Función 2
plt.plot(x, y2, color='black', marker='s', linestyle=':', linewidth=4, label='$-x^2 + 2$')
# Colocamos la leyenda abajo a la izquierda
plt.legend(loc='lower left')

# Mostramos el gráfico
plt.show()
````

````python
# Función 1
plt.plot(x, y1, 'ro-', label='$x^2 - 2$')
# Función 2
plt.plot(x, y2, color='black', marker='s', linestyle=':', linewidth=4, label='$-x^2 + 2$')
# Colocamos la leyenda arriba en el centro
plt.legend(loc='upper center')

# Mostramos el gráfico
plt.show()
````

## <font color='steelblue'> 4. Varios gráficos en un lienzo

En ocasiones, en lugar de superponer dos representaciones para compararlas, preferimos presentarlas en gráficos/ventanas distintas, pero visualizadas a la vez en el mismo lienzo. Para ello hemos de utilizar la función `subplots()` del módulo `matplotlib`.


Esta función tiene tres argumentos:
- el primero indica el número de **filas** del lienzo en el que ubicaremos los diferentes gráficos,
- el segundo argumento indica el número de **columnas**
- el tercero indica la posición que pretendemos para el gráfico a representar a continuación con `plot()`. Se trata de un número entero que marca el orden en la secuencia de gráficos que se suceden, recorriendo filas y columnas de izquierda a derecha, desde la esquina superior izquierda del lienzo hacia la inferior derecha.

A continuación del comando `subplot()`, dibujamos el gráfico correspondiente con `plot()`. Repetimos el proceso hasta ubicar todos los gráficos que queremos presentar en el mismo lienzo.

Veamos ahora cómo lo hacemos.

````python
# Una fila y dos columnas: subplot(1,2, )

# primer subgráfico
plt.subplot(1,2,1)
plt.plot(x, y1, 'ro-')

# segundo subgráfico
plt.subplot(1,2,2)
plt.plot(x, y2, 'ks:')

plt.show()
````

````python
# Dos filas y una columna: subplot(2,1, )

# primer subgráfico
plt.subplot(2,1,1)
plt.plot(x, y1, 'ro-')

# segundo subgráfico
plt.subplot(2,1,2)
plt.plot(x, y2, 'ks:')

plt.show()
````

También se pueden crear figuras con más subgráficos, es decir, con más filas y más columnas.

Como ejemplo, incorporamos dos funciones más: $y3=x^2 + 5$ e $y4=-x^2 - 5$, para lo que hemos de crear primero las nuevas listas con los valores de los resultados de las funciones.

````python
# creamos las "y"s vacías
y3 = []
y4 = []
for i in x:
  # añadimos los valores
  y3.append(i**2 + 5)
  y4.append(-i**2 - 5)

# verificamos las listas
print(x)
print(y3)
print(y4)
````

Ya podemos representar:

````python
# Dos filas y dos columnas: subplot(2,2,)

# primer subgráfico
plt.subplot(2,2,1)
plt.plot(x, y1, 'ro-')

# segundo subgráfico
plt.subplot(2,2,2)
plt.plot(x, y2, 'ks:')

# tercer subgráfico
plt.subplot(2,2,3)
plt.plot(x, y3, 'g^-.')

# cuarto subgráfico
plt.subplot(2,2,4)
plt.plot(x, y4, 'b*')

plt.show()
````

````python
# Una fila y cuatro columnas: subplot(1,4, )

# primer subgráfico
plt.subplot(1,4,1)
plt.plot(x, y1, 'ro-')

# segundo subgráfico
plt.subplot(1,4,2)
plt.plot(x, y2, 'ks:')

# tercer subgráfico
plt.subplot(1,4,3)
plt.plot(x, y3, 'g^-.')

# cuarto subgráfico
plt.subplot(1,4,4)
plt.plot(x, y4, 'b*')

plt.show()
````

La distribución de los gráficos no tiene por qué ser simétrica y podemos dejar huecos:

````python
# Dos filas y tres columnas: subplot(2,3, )

# primer subgráfico
plt.subplot(2,3,1)
plt.plot(x, y1, 'ro-')

# segundo subgráfico
plt.subplot(2,3,2)
plt.plot(x, y2, 'ks:')

# tercer subgráfico
plt.subplot(2,3,3)
plt.plot(x, y3, 'g^-.')

# cuarto subgráfico
plt.subplot(2,3,6)
plt.plot(x, y4, 'b*')

plt.show()
````

````python
# Dos filas y tres columnas: subplot(2,3, )

# primer subgráfico
plt.subplot(2,3,1)
plt.plot(x, y1, 'ro-')

# segundo subgráfico
plt.subplot(2,3,2)
plt.plot(x, y2, 'ks:')

# tercer subgráfico
plt.subplot(2,3,5)
plt.plot(x, y3, 'g^-.')

# cuarto subgráfico
plt.subplot(2,3,6)
plt.plot(x, y4, 'b*')

plt.show()
````

### <font color='steelblue'>Gráficos dentro de gráficos

En ocasiones nos interesa anidar (encajar) un gráfico dentro de otro.
Para ello utilizaremos la función `axes()`, que utilizaremos antes de pintar con `plot()` cada uno de los gráficos a visualizar. Los argumentos que recibe en cada ejecución son las coordenadas del marco en el que deseamos encuadrar cada gráfico:

```plt.axes([a,b,c,d])```

* a: la posición *x* de la esquina inferior izquierda.
* b: la posición *y* de la esquina inferior izquierda.
* c: la posición *x*, respecto de *a*, de la esquina superior derecha.
* d: la posición *y*, respecto de *a*, de la esquina superior derecha.

Con estas cuatro coordenadas (dos puntos), se dibuja el contorno del gráfico. Los valores pueden ir desde 0 hasta 1 (ambos inclusive), pero por facilidad de computación recomendamos evitar los extremos.

Para integrar el gráfico interior, hemos de asegurarnos que sus coordenadas son menores a las del gráfico contenedor. Este gráfico que se cree, se superpondrá al que ya existe.

Tras cada ejecución de `axes()`, ejecutamos el `plot()` correspondiente, y por último, como siempre, mostraremos el gráfico con `show()`.

Veamos un ejemplo con *y1* e *y2*:

````python
plt.axes([0.05,0.05,0.95,0.95])
plt.plot(x, y2, 'ro-')

plt.axes([0.25,0.25,0.35,0.35])
plt.plot(x, y1, 'ks:')

plt.show()
````

## <font color='steelblue'> 5. Configuración del entorno gráfico

Vemos a continuación cómo configurar otros elementos y características que conforman el entorno gráfico que mostramos al representar los objetos matemáticos que nos interesan.

### <font color='steelblue'>Etiquetas para título y ejes

Empezaremos con las funciones `title()`, `xlabel()` e `ylabel()`, con las que especificamos, respectivamente, el título del gráfico, el del eje de abscisas (eje *X*), y el del eje de ordenadas (eje *Y*). El argumento que reciben estas etiquetas es una cadena de texto, y por tanto irá entrecomillada.

Además de la etiqueta, podemos definir otras opciones adicionales, como la fuente, el tamaño de letra, ubicación, etc, que podemos consultar en la [ayuda de matplotlib](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.title.html).

Empezamos dando un título a la función *y1*: "Primera función".

````python
plt.plot(x, y1, 'ro-')
# ponemos el título
plt.title('Primera función')

plt.show()
````

Utilizamos ahora `fontsize` para el tamaño de letra y `loc` para la colocación:

### <font color='steelblue'>Amplitud de los ejes

En ocasiones queremos enfocar un gráfico aplicando un zoom de aumento o de decremento. Esto lo conseguimos modificando el rango de valores $x$ e $y$ que visualizamos, y lo hacemos en Pytnon con las funciones `xlim()` e `ylim()`, respectivamete para el eje *X* y el eje *Y*. Podemos indicar como argumentos el mínimo valor a visualizar (`xmin=`,`ymin=`), el máximo (`xmax=`,`ymax=`) o ambos (en cuyo caso, no es preciso indicar a qué argumento nos referimos, pero sí utilizar el orden (min,max).

Como ejemplo, queremos visualizar en el gráfico de `y1` los valores del 0 al 4 en el eje *X* y como valor máximo en el eje *Y* el 10.

````python
plt.plot(x, y1, 'ro-')
plt.title('La fórmula es: $y = x^2 - 2$', fontsize=20)
plt.xlabel('Eje x', fontsize=10)
plt.ylabel('Eje y', fontsize=10)
# delimitador de rango en el eje x
plt.xlim(0,4)
# delimitador de rango en el eje y
plt.ylim(ymax=10)

plt.show()
````

**Nota:** Si escribimos los límites en orden inverso, (max,min), el gráfico se invierte (al ir de mayor a menor), como si viéramos su reflejo en un espejo.

Al dibujar con la configuración por defecto, `plot` consigue un encuadre perfecto de los ejes (equivalente más o menos, a multiplicar por 1.1 los valores mínimos y máximo del rango de valores definidos).

````python
plt.plot(x, y1, 'ro-')
plt.title('La fórmula es: $y = x^2 - 2$', fontsize=20)
plt.xlabel('Eje x', fontsize=10)
plt.ylabel('Eje y', fontsize=10)
# delimitador de rango en el eje x
plt.xlim(min(x)*1.1, max(x)*1.1)
#delimitador de rango en el eje y
plt.ylim(min(y1)*1.1, max(y1)*1.1)

plt.show()
````

###  <font color='steelblue'> Marcas de los ejes

Podemos también indicar qué marcas y qué números/etiquetas queremos que aparezcan en cada uno de los ejes. Para ello utilizaremos las funciones `xticks()` e `yticks()`, que han de recibir como argumento principal una lista con los valores en los que queremos mostrar marcas.

Queremos visualizar, por ejemplo que solo aparezcan marcas en el eje $X$ en los valores $(-4,0,4)$ y en el eje *Y* en $(0,10,20)$.

````python
plt.plot(x, y1, 'ro-')
plt.title('La fórmula es: $y = x^2 - 2$', fontsize=20)
plt.xlabel('Eje x', fontsize=10)
plt.ylabel('Eje y', fontsize=10)
plt.xlim(min(x)*1.1, max(x)*1.1)
plt.ylim(min(y1)*1.1, max(y1)*1.1)
# eje x
plt.xticks([-4,0,4])
# eje y
plt.yticks([0,10,20])

plt.show()
````

También podemos definir nuestras propias etiquetas para cada una de las marcas, incluyendo como segundo argumento una lista con las etiquetas que deseamos, bien sean textos o expresiones matemáticas entrecomilladas.

Tocamos las etiquetas de las marcas a continuación.

````python
plt.plot(x, y1, 'ro-')
plt.title('La fórmula es: $y = x^2 - 2$', fontsize=20)
plt.xlabel('Eje x', fontsize=10)
plt.ylabel('Eje y', fontsize=10)
plt.xlim(min(x)*1.1, max(x)*1.1)
plt.ylim(min(y1)*1.1, max(y1)*1.1)
# eje x
plt.xticks([-4,0,4], ['-4m', '0m', '4m'])
# eje y
plt.yticks([0,10,20], ['$0m^2$', '$10m^2$', '$20m^2$'])

plt.show()
````

### <font color='steelblue'>Espacios entre gráficos

Cuando añadimos etiquetas a los ejes y títulos a los gráficos y los mostramos agrupados en un mismo lienzo, nos puede pasar algo como lo que vemos en la siguiente figura: que se superponen títulos y etiquetas y no se pueden leer.

````python
plt.subplot(2,2,1)
plt.plot(x, y1, 'ro-')
plt.title('Primera función')
plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.subplot(2,2,2)
plt.plot(x, y2, 'ks:')
plt.title('Segunda función')
plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.subplot(2,2,3)
plt.plot(x, y3, 'g^-.')
plt.title('Tercera función')
plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.subplot(2,2,4)
plt.plot(x, y4, 'b*')
plt.title('Cuarta función')
plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.show()
````

Para evitar este inconveniente tenemos como opción ampliar automáticamente el espacio entre los gráficos con la función `tight_layout()`. Esta función ajusta automáticamente los espacios entre gráficos para no solapar etiquetas, se utiliza después de crear todos los gráficos a ubicar en el mismo lienzo, y siempre antes de mostrarlo con `show()`:

````python
plt.subplot(2,2,1)
plt.plot(x, y1, 'ro-')
plt.title('Primera función')
plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.subplot(2,2,2)
plt.plot(x, y2, 'ks:')
plt.title('Segunda función')
plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.subplot(2,2,3)
plt.plot(x, y3, 'g^-.')
plt.title('Tercera función')
plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.subplot(2,2,4)
plt.plot(x, y4, 'b*')
plt.title('Cuarta función')
plt.xlabel('Eje x')
plt.ylabel('Eje y')

# ensanchamos
plt.tight_layout()

plt.show()
````

### <font color='steelblue'>Tamaño de un lienzo

Hasta ahora hemos comprobado cómo la función `show()` genera una figura o lienzo siempre con el mismo tamaño, independientemente de los gráficos que contenga, y adaptando el tamaño de estos al lienzo. Si queremos modificar el tamaño por defecto del lienzo o figura final, podemos hacerlo con la función `figure()`, que recibe dos argumentos:

* `figsize`: una tupla que indica el tamaño (ancho y alto) del lienzo en pulgadas (por defecto utiliza 6.4x4.8).
* `dpi`: un entero que indica el número de puntos por pulgada (por defecto utiliza 100), y hace referencia al tamaño del contenido en pantalla.

Esta función se ejecuta antes de crear todos los gráficos.

Con este módulo podemos utilizar diversos estilos para los gráficos. Para seleccionar el estilo hemos de ejecutar:

```
with plt.style.context(estilo):
```

antes de crear los gráficos, y utilizando como argumento de 'estilo' alguna de las siguientes opciones en forma de lista:

* `["classic"]`: genera un entorno gráfico clásico
* `["ipynb"]`: prescinde del borde de la leyenda y las marcas de los ejes son unidades, de modo que añade potencias de 10 para ajustar la escala y las referencia en los extremos de los ejes.
* `["ipynb","use_mathtext"]`: igual que el anterior, pero las potencias de 10 las expresa con notación científica.
* `["ipynb","use_mathtext","colors5-light"]`: igual que el anterior y además cambia el color de cada una de las representaciones que se incluyan en un gráfico según la paleta de 5 colores '*light*'.
* `["ipynb","use_mathtext","colors10-ls"]`: igual que el tercero y además cambia el estilo de la línea de cada una de las representaciones que se incluyan en el mismo gráfico.
* `["ipynb","use_mathtext","colors10-markers"]`: igual que el tercero, además cambia el estilo de los marcadores de cada una de las representaciones que se incluyan en el mismo gráfico.


Para mostrar los diferentes estilos utilizamos los datos de la [sección 3](https://colab.research.google.com/drive/1TP8lj4YDBZKpsePrQ_BUN4Lo1EwkeZKS#scrollTo=EjT3mXJ-yGVd&line=1&uniqifier=1) de este cuaderno.

````python
# Estilo clásico

# Carga de datos
x = list(range(-5,6))

# creamos las "y"s vacías
y1 = []
y2 = []
for i in x:
  # añadimos los valores
  y1.append(i**2 - 2)
  y2.append(-i**2 + 2)

# tamaño de la figura
plt.figure(figsize=(10,7), dpi=80)

with plt.style.context(["classic"]):
  # Función 1
  plt.plot(x, y1,label='$x^2 - 2$')
  # Función 2
  plt.plot(x, y2, label='$-x^2 + 2$')
  # Indicamos la leyenda
  plt.legend()

  # Mostramos el gráfico
  plt.show()
````

````python
# Escala de los ejes en potencias de 10
plt.figure(figsize=(10,7), dpi=80)

with plt.style.context(["ipynb"]):
  # Función 1
  plt.plot(x, y1,label='$x^2 - 2$')
  # Función 2
  plt.plot(x, y2, label='$-x^2 + 2$')
  # Indicamos la leyenda
  plt.legend()

  # Mostramos el gráfico
  plt.show()
````

````python
# Escala de los ejes en potencias de 10 con notación científica
plt.figure(figsize=(10,7), dpi=80)

with plt.style.context(["ipynb","use_mathtext"]):
  # Función 1
  plt.plot(x, y1,label='$x^2 - 2$')
  # Función 2
  plt.plot(x, y2, label='$-x^2 + 2$')
  # Indicamos la leyenda
  plt.legend()

  # Mostramos el gráfico
  plt.show()
````

````python
# Escala de los ejes en potencias de 10 con notación científica y paleta 5light
plt.figure(figsize=(10,7), dpi=80)

with plt.style.context(["ipynb","use_mathtext","colors5-light"]):
  # Función 1
  plt.plot(x, y1,label='$x^2 - 2$')
  # Función 2
  plt.plot(x, y2, label='$-x^2 + 2$')
  # Indicamos la leyenda
  plt.legend()

  # Mostramos el gráfico
  plt.show()
````

````python
# Escala de los ejes en potencias de 10 con notación científica, paleta 10 y distintos tipos de líneas
plt.figure(figsize=(10,7), dpi=80)

with plt.style.context(["ipynb","use_mathtext","colors10-ls"]):
  # Función 1
  plt.plot(x, y1,label='$x^2 - 2$')
  # Función 2
  plt.plot(x, y2, label='$-x^2 + 2$')
  # Indicamos la leyenda
  plt.legend()

  # Mostramos el gráfico
  plt.show()
````

````python
# Escala de los ejes en potencias de 10 con notación científica, paleta 10 y distintos tipos de marcadores
plt.figure(figsize=(10,7), dpi=80)

with plt.style.context(["ipynb","use_mathtext","colors10-markers"]):
  # Función 1
  plt.plot(x, y1,label='$x^2 - 2$')
  # Función 2
  plt.plot(x, y2, label='$-x^2 + 2$')
  # Indicamos la leyenda
  plt.legend()

  # Mostramos el gráfico
  plt.show()
````
